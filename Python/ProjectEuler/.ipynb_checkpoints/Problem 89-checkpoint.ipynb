```python
# Problem 89

from time import time

starttime = time()

lines = []
with open('p089_roman.txt') as f:
    for line in f:
        for l in line.strip().split(","):
            lines.append(l)
f.close()

def conv_number(r):
    if r == "I":
        return 1
    elif r == "V":
        return 5
    elif r == "X":
        return 10
    elif r == "L":
        return 50
    elif r == "C":
        return 100
    elif r == "D":
        return 500
    elif r == "M":
        return 1000
    else:
        return 0
    
def convert_to_number(roman_number : str):
    sum = 0
    
    for char in roman_number:
        sum += conv_number(char)
        
    return sum

def conv_roman(number : int, one : str, five : str):
    ret = ""
    
    if number == 0:
        pass
    elif number > 0 and number < 4:
        for i in range(0, number):
            ret += one
    elif number == 4:
        ret = one + five
    elif number == 5:
        ret = five
    else:
        ret = five
        for i in range(0, number - 5):
            ret += one
    
    return ret
    
    
def convert_to_roman(number):
    ret = ""
    
    i = len(str(number)) - 1
    
    for char in str(number):
        if i >= 3: # 4 Zeichen
            ret += conv_roman(int(char), list_roman[1000], list_roman[5000])
        elif i >= 2:
            ret += conv_roman(int(char), list_roman[100], list_roman[500])
        elif i >= 1:
            ret += conv_roman(int(char), list_roman[10], list_roman[50])
        elif i >= 0:
            ret += conv_roman(int(char), list_roman[1], list_roman[5])
        
        i -= 1
    
    return ret

list_roman = {5000: "v", 1000 : "M", 500 : "D", 100 : "C", 50 : "L", 10 : "X", 5 : "V", 1 : "I"}

sum = 0
old_sum = 0

for l in lines:
    #print(l, convert_to_number(l), convert_to_roman(convert_to_number(l)))
    sum += len(convert_to_roman(convert_to_number(l)))
    old_sum += len(l)
    
    
print("Time: {}, Sum: {}".format(time() - starttime, old_sum - sum))

import re
ret = "".join(lines)
replaced = re.sub('DCCCC|LXXXX|VIIII|CCCC|XXXX|IIII', 'aa', ret)
print("Time: {}, Sum: {}".format(time() - starttime, len(ret) - len(replaced)))
```
