```python
# Problem 61
import math
from sympy.solvers import solve
from sympy import *
import itertools
import time

def polygonal(p, n):   
    if p == 3:
        erg = n * (n+1)/2
    elif p == 4:
        erg = n**2
    elif p == 5:
        erg = n*(3*n-1)/2
    elif p == 6:
        erg = n*(2*n-1)
    elif p == 7:
        erg = n*(5*n-3)/2
    elif p == 8:
        erg = n*(3*n-2)
    else:
        pass
    return int(erg)

def reverse_polygonal(p, result):
    n = Symbol("n", positive=True)
    
    if p == 3:
        erg = solve(n * (n+1)/2 - result, n)
    elif p == 4:
        erg = math.sqrt(result)
    elif p == 5:
        erg = solve(n*(3*n-1)/2 - result, n)
    elif p == 6:
        erg = solve(n*(2*n-1)-result, n)
    elif p == 7:
        erg = solve(n*(5*n-3)/2-result, n)
    elif p == 8:
        erg = solve(n*(3*n-2)-result, n)
    else:
        pass
    return int(N(erg[0]))

def is_cyclic_to(a : str, b : str):
    if a[-2:] == b[:2]:
        return True
    else:
        return False

def get_type(type, number):
    erg = []
    for i in range(3, k + 2 + 1):
        if number in type[i]:
            erg.append(i)
            
    return erg
        
def same_type(type, number, i):
    return len(set(get_type(type, number)) - set(get_type(type, i))) == 0
    
starttime = time.time()

numbers = []
type = {}
k = 6

for p in range(3, k + 2 + 1):
    n = 0
    i = 1
    while n < 10000:
        n = polygonal(p, i)
        
        if n >= 1000 and n < 10000:
            numbers.append(n)
            try:
                type[p].append(n)
            except KeyError:
                type[p] = [n]
        i += 1
        
numbers = list(set(numbers))
numbers.sort()

#print(numbers)
         
chains = [[x] for x in numbers]
for round in range(1, k):
    for chain in chains:
        for number in numbers:
            if number not in chain and is_cyclic_to(str(chain[-1]), str(number)):
                if round == k-1 and not is_cyclic_to(str(number), str(chain[0])):
                    continue
                
                bool = True
                for i in chain:
                    if same_type(type, number, i):
                        bool = False
                        break

                if bool:
                    chain.append(number)
                    break

for chain in chains:
    if len(chain) >= k-1:       
        sum = 0
        for i in chain:
            sum += i
        
        print("Sum: {} for {}".format(sum, chain))
        
print("Needed time: {}".format(time.time() - starttime))
```
