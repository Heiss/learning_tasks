```python
# Problem 57

n = 10**3

# a / b
a = 1
b = 1

p = {}
p[0] = 0
p[1] = 1
count = 0

def pell_number(n):
    if n not in p:
        p[n] = 2 * pell_number(n - 1) + pell_number(n - 2)
    
    return p[n]

for i in range(1, n + 1):
    a = pell_number(i - 1) + pell_number(i)
    b = pell_number(i)
    
    if len(str(a)) > len(str(b)):
        count += 1
          
print(count)
```

```python
# Problem 58
from sympy.ntheory import isprime
from math import sqrt

count = 0
n = 1
e = 2
i = 1

ratio = 1

while ratio > 0.1:
    # iterate over the edges
    i += e
    
    # entered a new layer?
    if sqrt(i) % 2 == 1:
        e += 2
        n += 2
        
    if isprime(i):
        count += 1
        
    ratio = count / (n * 2 -1)

print("Count: {}, Diagonal: {}, Ratio: {}".format(count, n * 2 -1, ratio))
print("Side length: {}".format(n))
```

```python
# Problem 60
import sys
from sympy.ntheory import isprime
from sympy import sieve
import time

def concat(primes : list):
    list = []
    
    for i in primes:
        for j in primes:
            if i != j:
                list.append(int(str(i)+str(j)))
    
    return list
                        
def all_primes(check : list):
    bool = True
    
    for i in check:
        if not isprime(i):
            bool = False
        
    return bool

print("Prime generating at {}".format(time.time()))
highest_prime = 10**4 # highest prime
count_of_pairs = 5 # count of pairs

sieve.extend(highest_prime)
primes = sieve._list
        
counter = 0
pairs = {}

starttime = time.time()
print("First walkthrough")
# first walkhrough to generate the pairs, which are possible
for idx, p in enumerate(primes):
    l1 = []

    for i in range(0, idx):
    #for current in primes:
        current = primes[i]
        if p != current and all_primes(concat([p, current])):
           l1.append(current) 
    if len(l1) + 1 > count_of_pairs:
        pairs[(p,)] = l1

snd_starttime = time.time()
smallest_sum = sys.maxsize

print("Start...")
while True:
    if counter % 1000 == 0:
        print("Working...")
    try:
        key, value = pairs.popitem()
    except KeyError:
        print("All done.")
        break
    
    if len(value) == 0:
        if len(key) == count_of_pairs:
            sum = 0
            for i in key:
                sum += i
            if sum < smallest_sum:
                smallest_sum = sum
                
        continue
    
    first_number = value.pop(0)
        
    new_key = list(key)
    new_key.append(first_number)
    
    
    l1 = []
    for p in value:
        if all_primes(concat([first_number, p])):
              l1.append(p)
                
    if len(new_key) + len(l1) >= count_of_pairs:
        pairs[tuple(new_key)] = l1
    counter += 1

endtime = time.time()
print("time: {}, working: {}".format(endtime - starttime, endtime - snd_starttime))
print("smallest sum: {}".format(smallest_sum))
```

```python
# Problem 61
import math
from sympy.solvers import solve
from sympy import *
import itertools
import time

def polygonal(p, n):   
    if p == 3:
        erg = n * (n+1)/2
    elif p == 4:
        erg = n**2
    elif p == 5:
        erg = n*(3*n-1)/2
    elif p == 6:
        erg = n*(2*n-1)
    elif p == 7:
        erg = n*(5*n-3)/2
    elif p == 8:
        erg = n*(3*n-2)
    else:
        pass
    return int(erg)

def reverse_polygonal(p, result):
    n = Symbol("n", positive=True)
    
    if p == 3:
        erg = solve(n * (n+1)/2 - result, n)
    elif p == 4:
        erg = math.sqrt(result)
    elif p == 5:
        erg = solve(n*(3*n-1)/2 - result, n)
    elif p == 6:
        erg = solve(n*(2*n-1)-result, n)
    elif p == 7:
        erg = solve(n*(5*n-3)/2-result, n)
    elif p == 8:
        erg = solve(n*(3*n-2)-result, n)
    else:
        pass
    return int(N(erg[0]))

def is_cyclic_to(a : str, b : str):
    if a[-2:] == b[:2]:
        return True
    else:
        return False

def get_type(type, number):
    erg = []
    for i in range(3, k + 2 + 1):
        if number in type[i]:
            erg.append(i)
            
    return erg
        
def same_type(type, number, i):
    return len(set(get_type(type, number)) - set(get_type(type, i))) == 0
    
starttime = time.time()

numbers = []
type = {}
k = 6

for p in range(3, k + 2 + 1):
    n = 0
    i = 1
    while n < 10000:
        n = polygonal(p, i)
        
        if n >= 1000 and n < 10000:
            numbers.append(n)
            try:
                type[p].append(n)
            except KeyError:
                type[p] = [n]
        i += 1
        
numbers = list(set(numbers))
numbers.sort()

#print(numbers)
         
chains = [[x] for x in numbers]
for round in range(1, k):
    for chain in chains:
        for number in numbers:
            if number not in chain and is_cyclic_to(str(chain[-1]), str(number)):
                if round == k-1 and not is_cyclic_to(str(number), str(chain[0])):
                    continue
                
                bool = True
                for i in chain:
                    if same_type(type, number, i):
                        bool = False
                        break

                if bool:
                    chain.append(number)
                    break

for chain in chains:
    if len(chain) >= k-1:       
        sum = 0
        for i in chain:
            sum += i
        
        print("Sum: {} for {}".format(sum, chain))
        
print("Needed time: {}".format(time.time() - starttime))
```

```python
# Problem 62
import itertools
import math
import sys
import time

starttime = time.time()

print("Start working...")
for i in range(111, 10**6):
    if i % 10**4 == 0:
        print("working...")
    
    counter = 0
    smallest_cube = sys.maxsize
    breaker = None
    
    c = i**3
    
    l1 = list(map("".join, itertools.permutations(str(c))))
    
    for l in l1:
        if int(l) < 111:
            continue
            
        cube = float(l)**(1/3.0)
        if cube.is_integer():
            if cube < smallest_cube:
                smallest_cube = cube
            counter += 1
        
        if counter == 3:
            breaker = True
            break
    if breaker is not None:
        break

print("Timer: {}, cube: {}".format(time.time() - starttime, smallest_cube))
        
```
