```python
# Problem 81
import time
import numpy as np
import sys

starttime = time.time()

target=80
lines = []
with open('p081_matrix.txt') as f:
    for line in f:
        for l in line.strip().split(","):
            lines.append(int(l))
f.close()

arr = np.array(lines)
mat = arr.reshape((target,target))

for i in range(0, target):
    for j in range(0, target):
        smallest_integer = sys.maxsize
        
        if i == 0 and j == 0:
            continue
        
        if i > 0:
            ret = mat[i,j] + mat[i-1, j]
            if ret < smallest_integer:
                smallest_integer = ret
        if j > 0:
            ret = mat[i,j] + mat[i, j-1]
            if ret < smallest_integer:
                smallest_integer = ret
                
        mat[i,j] = smallest_integer

print("Time: {}, Smallest sum: {}".format(time.time() - starttime, mat[-1,-1]))
```

```python
# Problem 82
import time
import numpy as np
import sys

starttime = time.time()

target=80
lines = []
with open('p081_matrix.txt') as f:
    for line in f:
        for l in line.strip().split(","):
            lines.append(int(l))
f.close()

arr = np.array(lines)
mat = arr.reshape((target,target))
path = mat.copy()

# go through columns from right to left
for j in range(target-2, -1, -1): 
    # right and top right
    for i in range(0,target):
        # right
        path[i,j] = mat[i,j+1] + mat[i,j]
        # top + top right
        if i > 0:
            ret = path[i-1,j] + mat[i,j]
            if ret < path[i,j]:
                path[i,j] = ret
        
    # down right
    for i in range(target-2, -1, -1):
        # down + down right
        if i < target-2:
            ret = path[i + 1,j] + mat[i,j]
            if ret < path[i,j]:
                path[i,j] = ret 
        
    mat = path.copy()
print("Time: {}, Smallest sum: {}".format(time.time() - starttime, np.min(mat[:,0])))
```

```python
# Problem 83
# Dijkstra algorithm - worktime -.- too long! 6 Hours

import time
import numpy as np
import sys

starttime = time.time()

target=80
lines = []
with open('p081_matrix.txt') as f:
    for line in f:
        for l in line.strip().split(","):
            lines.append(int(l))
f.close()

arr = np.array(lines)
mat = arr.reshape((target,target))
G = np.ones((target,target))*sys.maxsize
G[0,0] = mat[0,0]
    
#Startpoint
open_list = [(0,0)]
last_node = (0,0)

parent = {}
for i in range(target):
    for j in range(target):
        parent[(i,j)] = None
open_list = list(parent.keys())

while len(open_list) > 0:
    # take the node with lowest value
    current_node = open_list[0]
    save = 0
    for idx, node in enumerate(open_list):
        if G[node] < G[current_node]:
            current_node = node
            save = idx
    current_node = open_list.pop(save)
       
    # all new nodes which should be visit in the next round
    new_node = []
    
    # top
    if current_node[1] > 0:
        new_node.append((current_node[0], current_node[1]-1))

    # down
    if current_node[1] < target-1:
        new_node.append((current_node[0], current_node[1]+1))

    # left
    if current_node[0] > 0:
        new_node.append((current_node[0]-1, current_node[1]))

    # right
    if current_node[0] < target-1:
        new_node.append((current_node[0]+1, current_node[1]))

    # check all the conditions for the new nodes
    for new in new_node:
        if new in open_list:
            update = G[current_node] + mat[new]
            if update < G[new]:
                parent[new] = current_node
                G[new] = update

print("Time: {}, Last Node: {}".format(time.time() - starttime, G[-1,-1]))
```
