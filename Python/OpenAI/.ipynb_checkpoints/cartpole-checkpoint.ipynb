```python
import gym
import numpy as np
import sys
import time

def cartpole_rnn():
    total_reward = list()

    try:
        observation_dimension = 2
        input_dimension = 8
        hidden_dimension = 16
        output_dimension = 1

        learning_rate = 0.1
        epochs = 1000

        training_threshold = .3

        # weight values
        w0 = np.random.normal(0, 1, [input_dimension, hidden_dimension])
        w1 = np.random.normal(0, 1, [hidden_dimension, output_dimension])
        wh = np.random.normal(0, 2, [hidden_dimension, hidden_dimension])

        # delta values
        d0 = np.zeros_like(w0)
        d1 = np.zeros_like(w1)
        dh = np.zeros_like(wh)

        errs = list()
        accs = list()

        error = 0

        def sigmoid(x):
            return 1 / (1 + np.exp(-x))

        def deriv_sigmoid(out):
            return out * (1 - out)

        env = gym.make('CartPole-v0')
        for i_episode in range(epochs):
            start_time = time.time()


            done = False
            observation = env.reset()
            last_observation = np.zeros_like(observation)

            action = 0
            last_action = 0

            t = 0

            overall_err = 0 # total error in the whole calculation process.

            hidden = np.zeros(hidden_dimension)
            future_delta = np.zeros(hidden_dimension)

            partial_reward = 0.0


            while not done:
                env.render()

                norm = np.linalg.norm(observation)
                if norm > training_threshold:
                    action = np.abs(action - 1)

                X = np.array([np.concatenate((last_observation, observation), axis=0)]) # shape=(1, 8)
                Y = np.array([[action]]) # shape=(1, 1)

                # prediction calculation
                prev_hidden = hidden
                hidden = sigmoid(np.dot(X, w0) + np.dot(prev_hidden, wh))
                output = sigmoid(np.dot(hidden, w1))

                pred = np.round(output)[0].astype(np.int32).item()

                # squared mean error
                output_err = Y - output
                output_delta = output_err * deriv_sigmoid(output)
                overall_err += np.abs(output_err[0])

                # do sth with this result!
                last_action = action
                action = pred
                last_observation = observation
                observation, reward, done, info = env.step(action)

                # backpropagation through time     
                hidden_delta = ((np.dot(future_delta, wh.T) 
                                 + np.dot(output_delta, w1.T)) 
                                 * deriv_sigmoid(hidden))

                d1 += np.dot(np.atleast_2d(hidden).T, output_delta)
                dh += np.dot(np.atleast_2d(prev_hidden).T, hidden_delta)
                d0 += np.dot(X.T, hidden_delta)

                future_delta = hidden_delta

                w1 += learning_rate * d1
                w0 += learning_rate * d0
                wh += learning_rate * dh

                d1 *= 0
                d0 *= 0
                dh *= 0

                error += overall_err

                if done:
                    elapsed_time = time.time() - start_time

                    #print('Episode %d finished after %d timesteps. Total reward: %1.0f. Elapsed time: %d s' % (i_episode+1, t+1, partial_reward, elapsed_time))
                    total_reward.append(partial_reward)
                    break
                t += 1
                partial_reward += 1


    finally:
        env.render(close=True)
        arr = total_reward[-100:-1]
        print('Average reward: %3.2f' % np.mean(arr)),
        plt.plot(total_reward)


if __name__ == "__main__":
    cartpole_rnn()
```

```python
%matplotlib inline

import numpy as np
import gym, time
import collections
from lstm import LstmParam, LstmNetwork    
from heapq import heappush, heappop, heapreplace
import warnings
warnings.simplefilter("error", RuntimeWarning)
import matplotlib
import matplotlib.pyplot as plt

class memory_node:
    def __init__(self, reward, observations, actions):
        self.reward = reward
        self.observations = observations
        self.actions = actions
        
    def __lt__(self, other):
        if self.reward < other.reward:
            return True
        return False
    
    def __str__(self):
        return "{}, {}, {}".format(self.reward, self.observations, self.actions)

    
    
class cartpole_lstm_functions:
    """
    Computes square loss with first element of hidden layer array.
    """
    @classmethod
    def loss(self, pred, label):
        return (pred[-1] - label) ** 2

    @classmethod
    def bottom_diff(self, pred, label):
        diff = np.zeros_like(pred)
        diff[0] = 2 * (pred[0] - label)
        return diff

def cartpole_lstm():
    total_reward = list()

    try:
        epochs = 1000
        training_norm_threshold = .5
        learning_rate = .1
        mem_cell_ct = 128
      
        x_dim = 4
        max_frames = 16
        max_memory = 1
        
        lstm_param = LstmParam(mem_cell_ct, x_dim)
        lstm_net = LstmNetwork(lstm_param)
        
        i = 0
        memorys = list()
        for _ in range(max_memory):
            heappush(memorys, memory_node(i, [[0,0,0,0]], [[0]]))
            i -= 1
        
        env = gym.make('CartPole-v0')
        for i_episode in range(epochs):
            start_time = time.time()
            done = False
            t = 0
            partial_reward = 0.0
            i = 0

            observation = env.reset()
            x_list = collections.deque([], maxlen=max_frames)
            y_list = collections.deque([], maxlen=max_frames)
            
            y_list.extend(np.random.randint(2, size=max_frames))
            for _ in range(max_frames):
                x_list.append(observation)

            lstm_net.x_list_clear()

            while not done:
                env.render()

                for x in x_list:
                    lstm_net.x_list_add(np.asarray(x))

                action = np.clip(np.round(lstm_net.lstm_node_list[-1].state.h[0]).astype(np.int8), 0, 1).item()
                x_list.append(observation)
                observation, reward, done, info = env.step(action)
                
                if done:
                    elapsed_time = time.time() - start_time

                    #print('Episode %d finished after %d timesteps. Total reward: %1.0f. Elapsed time: %d s' % (i_episode+1, t+1, partial_reward, elapsed_time))
                    total_reward.append(partial_reward)
                    
                    # save the best reward
                    mem_node = memory_node(partial_reward, x_list, y_list)
                    check = heappop(memorys)
                    if(check.reward > partial_reward):
                        heappush(memorys, check)
                    else:
                        heappush(memorys, mem_node)

                    break
                
                t += 1
                partial_reward += reward
                
                #was our action a good choice?
                norm = np.linalg.norm(observation)
                if norm > training_norm_threshold:
                    action = np.abs(action - 1)
                
                y_list.append(action)

                loss = lstm_net.y_list_is(y_list, cartpole_lstm_functions)
                lstm_param.apply_diff(lr=learning_rate)
                lstm_net.x_list_clear()
            
            """# train the best rewards
            lstm_net.x_list_clear()
            mean = np.mean([node.reward for node in memorys])
            
            for node in memorys:
                if (node.reward < mean) or (node.reward < 10):
                    continue
                    
                for x in node.observations:
                    lstm_net.x_list_add(np.asarray(x))
                
                loss = lstm_net.y_list_is(node.actions, cartpole_lstm_functions)
                lstm_param.apply_diff(lr=learning_rate)
                lstm_net.x_list_clear()
                
            
            if len(total_reward) > 1:
                print(mean, np.mean(total_reward[-100:-1]), loss)"""
    
    finally:
        #env.render(close=True)
        
        arr = total_reward[-100:-1]
        print('Average reward: %3.2f' % np.mean(arr))
        plt.plot(total_reward)
        
if __name__ == "__main__":
    cartpole_lstm()
```
